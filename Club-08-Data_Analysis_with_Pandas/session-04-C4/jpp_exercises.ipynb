```python
# Question 3
df = pd.read_csv('exercises/faang.csv', index_col='date', parse_dates=['date'])
df.head()
```

```python
import numpy as np
df.groupby(['ticker']).resample("M").agg({
        'open': np.mean,
        'high': np.max,
        'low': np.min,
        'close': np.mean,
        'volume': np.sum,
})
```

```python
# Question 4
df = pd.read_csv('exercises/faang.csv', index_col='date', parse_dates=['date'])
df.head()
```

```python
# Question 7
df = pd.read_csv('exercises/faang.csv', index_col='date', parse_dates=['date'])
df.head()
```

```python
from scipy.stats import zscore as zscore
nflx = df.query('ticker == "NFLX"')
```

```python
nflx[['open', 'high', 'low', 'close', 'volume']].apply(
    lambda x: x.sub(x.mean()).div(x.std())
)
```

```python
# Question 8
df_event = pd.DataFrame({"ticker" : "FB", 
                         "date" : ['2018-07-25', '2018-03-19', '2018-03-20'],
                         "event" : ["Disapointing user growth announced after close",
                                    "Cambridge Analytica Story", "FTC Investigation"]},
            ).set_index(['ticker', 'date'])
df = pd.read_csv('exercises/faang.csv', index_col='date', parse_dates=['date'])
```

```python
pd.merge(df.reset_index(drop=False).set_index(['ticker', 'date']),
         df_event, left_index=True, right_index=True, how='outer')
```
