```python
import pandas as pd
```

```python
# Question 1
stock = ['aapl', 'amzn', 'fb', 'goog', 'nflx']
file = []
for name in stock:
    tmp = pd.read_csv('exercises/' + name + '.csv')
    tmp['ticker'] = name.upper()
    file.append(tmp)
df = pd.concat(file).reset_index(drop = True)
df.to_csv('exercises/faang.csv', index = False)
```

```python
# Question 2
df = df.astype({'volume': 'int32'})
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
df.sort_values(by=['date'])
```

```python
# Question 3
df.sort_values(by=['volume'], ascending=False).iloc[1:7, :]
```
